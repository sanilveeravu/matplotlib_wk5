# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
import matplotlib.ticker as mtick

# File to Load (Remember to change these)
city_data_to_load = os.path.join("Resources","city_data.csv")
ride_data_to_load = os.path.join("Resources","ride_data.csv")

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_by_city_type = pyber_data_df.groupby("type").count()["ride_id"]
total_rides_by_city_type
```

```python
# 2. Get the total drivers for each city type
total_drivers_by_city_type = city_data_df.groupby("type").sum()["driver_count"]
total_drivers_by_city_type
```

```python
#  3. Get the total amount of fares for each city type
total_fares_by_city_type = pyber_data_df.groupby("type").sum()["fare"]
total_fares_by_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride = total_fares_by_city_type / total_rides_by_city_type
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = total_fares_by_city_type / total_drivers_by_city_type
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame()
pyber_summary_df["Total Rides"]=total_rides_by_city_type
pyber_summary_df["Total Drivers"]=total_drivers_by_city_type
pyber_summary_df["Total Fares"]=total_fares_by_city_type
pyber_summary_df["Average Fare per Ride"]=average_fare_per_ride
pyber_summary_df["Average Fare per Driver"]=average_fare_per_driver
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df_formatted = pyber_summary_df.copy()

pyber_summary_df_formatted["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df_formatted["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df_formatted["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df_formatted["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df_formatted["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
pyber_summary_df_formatted
```

```python
# Extra Stuff
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4))
fig.suptitle('Rides and Drivers Distribution by City Type')
labels= ["Rural","Suburban","Urban"]
colors= ["Yellow","Seagreen","Firebrick"]
ax1.pie(total_rides_by_city_type, labels = labels, colors = colors, autopct='%.2f%%')
ax1.set_xlabel("Rides")
ax2.pie(total_drivers_by_city_type, labels = labels, colors = colors, autopct='%.2f%%')
ax2.set_xlabel("Drivers")
plt.savefig("analysis/CityTypeRidesDrivers.png")

plt.show()
```

```python
# Extra Stuff

df = pd.DataFrame(np.random.rand(10, 3),
                  columns=['a', 'b', 'c'])

fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(14,6))

colors= ["Yellow","Seagreen","Firebrick"]

total_fares_by_city_type.plot(ax=axes[0], kind='bar', 
                              title="Total Fares", xlabel="",color=colors)
average_fare_per_ride.plot(ax=axes[1], kind='bar', 
                           title="Avg. Fare Per Ride", xlabel="",color=colors);
average_fare_per_driver.plot(ax=axes[2], kind='bar', 
                             title="Avg. Fare Per Driver", xlabel="",color=colors);

fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
axes[0].yaxis.set_major_formatter(tick) 
axes[1].yaxis.set_major_formatter(tick) 
axes[2].yaxis.set_major_formatter(tick) 

plt.savefig("analysis/CityTypeFares.png")

plt.show()
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fare_by_type_date_df = pd.DataFrame(pyber_data_df.groupby(["type","date"]).sum()["fare"])
fare_by_type_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
fare_by_type_date_df = fare_by_type_date_df.reset_index()
fare_by_type_date_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fare_by_type_date_df_pivoted = fare_by_type_date_df.pivot(index='date', columns='type', values='fare')
fare_by_type_date_df_pivoted.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
date_mask=(fare_by_type_date_df_pivoted.index>='2019-01-01') & (fare_by_type_date_df_pivoted.index<'2019-04-29')
fare_by_type_date_df_for_date_range=fare_by_type_date_df_pivoted.loc[date_mask]
fare_by_type_date_df_for_date_range
```

```python
# Check data type before conversion
fare_by_type_date_df_for_date_range.info()
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
fare_by_type_date_df_for_date_range.index = pd.to_datetime(fare_by_type_date_df_for_date_range.index)
fare_by_type_date_df_for_date_range
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
fare_by_type_date_df_for_date_range.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fare_by_type_date_df_resampled = fare_by_type_date_df_for_date_range.resample('W').sum()
fare_by_type_date_df_resampled
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
#style.use('Solarize_Light2')

fig, ax = plt.subplots(1, 1, figsize=(18,6))

fare_by_type_date_df_resampled.plot(
    kind='line',
    xlabel="",
    ylabel="Fare ($USD)",
    title="Total Fare by City Type",
    ax=ax
)

fig.savefig("analysis/PyBer_fare_summary.png")

plt.show()
```
